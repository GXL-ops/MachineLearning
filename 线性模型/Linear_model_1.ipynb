```python
# 一元线性回归
# 生成训练数据
import os
os.environ['TF_CPP_MIN_LOG_LEVEL']='2'
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
# 打印内容不限长度
# np.set_printoptions(threshold=np.nan)
# 生成x
t_x=np.linspace(-1,1,50,dtype=np.float32)
noise=np.random.normal(0,0.05,t_x.shape)
t_y=t_x*3.0+5.0+noise
plt.plot(t_x,t_y,'k.')
plt.show()
```

```python
learning_rate=0.001
# 定义训练模型
x=tf.placeholder(tf.float32)
y=tf.placeholder(tf.float32)
a=tf.Variable(0.0)
b=tf.Variable(0.0)
curr_y=x*a+b
loss=tf.reduce_sum(tf.square(curr_y-y))
optimizer=tf.train.GradientDescentOptimizer(learning_rate)
train=optimizer.minimize(loss)
```

```python
# 进行训练数据
# 总共训练1000次
training_epochs=1000
# 创建Session
sess_new=tf.Session()
# 变量初始化
sess_new.run(tf.global_variables_initializer())
for i in range(training_epochs):
    sess_new.run(train,{x:t_x,y:t_y})
    if i%20==0:
        print(i,sess_new.run([a,b,loss],{x:t_x,y:t_y}))
a_val=sess_new.run(a)
b_val=sess_new.run(b)
print("this model is y=",a_val,"x+",b_val)
sess_new.close()
y_learned=t_x*a_val+b_val
plt.plot(t_x,t_y,'k.')
plt.plot(t_x,y_learned,'g-')
plt.show()
plt.close()
```
